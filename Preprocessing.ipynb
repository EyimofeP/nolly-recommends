# Data Preprocessing
___

This notebook involves the collection of the data i.e. Nigerian movies from the Wikipedia site, the movies range from the year 2016 - 2022.

```python
# Importing necessary libraries
import pandas as pd
```

## 2016

```python
# link of 2016 wikipedia data
link = "https://en.wikipedia.org/wiki/List_of_Nigerian_films_of_2016"
```

```python
# Collecting the HTML tables with movies
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
```

```python
# Merging the data together
data = pd.concat([df1, df2], ignore_index=True)
data.head()
```

```python
data
```

```python
# Getting only th Movie Title, Director, Cast and Genre Columns
data = data[["Title", "Director", "Cast", "Genre"]]
```

```python
data_2016 = pd.read_csv("data/2016.csv")
```
